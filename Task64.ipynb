```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
im_src = cv2.imread('chess_1.png')
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_src)
plt.show
```

```python
pts_src = np.array([[197,243],[192,48],[460,40],[459,262]])
```

```python
im_dst = cv2.imread('chess_2.png')
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_dst)
plt.show
```

```python
pts_dst = np.array([[407,364],[188,315],[166,32],[531,103]])
```

```python
h, status = cv2.findHomography(pts_src, pts_dst)
h
```

```python
im_out = cv2.warpPerspective(im_src, h,
                            (im_dst.shape[1],
                            im_dst.shape[0]))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_out)
plt.show
```

```python
image = im_out

flip_image = cv2.flip(image,1)
cv2.imshow("Flip image", flip_image)
cv2.waitKey(0)
(h, w) = image.shape[:2]
center = (w / 2, h / 2)

# get final horizontal image
M = cv2.getRotationMatrix2D(center, 270, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
cv2.imshow("Rotated image", rotated)
cv2.waitKey(0)
```

```python
plt.figure(figsize=(12,8))
plt.imshow(rotated)
plt.show
```
